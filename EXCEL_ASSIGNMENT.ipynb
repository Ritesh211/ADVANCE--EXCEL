


Question 1
Explain the difference between Absolute, Relative, and Mixed Cell Referencing in Excel with examples.
Answer:
Cell referencing in Excel defines how a cell reference behaves when a formula is copied to another cell.

1. Relative Cell Reference
Definition: Changes automatically when the formula is copied.

Example:
Formula in C2:

=A2+B2
When copied to C3, it becomes:

=A3+B3
Use Case: Used when the same calculation is needed row-wise or column-wise.

2. Absolute Cell Reference
Definition: Does not change when the formula is copied.

Syntax: $A$1

Example:

=A2*$B$1
$B$1 remains fixed even when copied.

Use Case: Used for constant values like tax rate, discount rate, etc.

3. Mixed Cell Reference
Definition: Either row or column is fixed.

Types:

$A1 → Column fixed

A$1 → Row fixed

Example:

=$A2*B$1
Use Case: Useful in tables like multiplication tables or matrices.

Conclusion:
Type	Changes when copied	Example
Relative	Row & Column	A1
Absolute	None	$A$1
Mixed	One part	$A1 / A$1

Question 5
Define the purpose of VLOOKUP and HLOOKUP. How are they different from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best?
Answer:
VLOOKUP
Searches vertically

Looks from left to right

=VLOOKUP(A2,A1:C10,2,FALSE)
HLOOKUP
Searches horizontally

=HLOOKUP(A1,A1:D5,2,FALSE)
XLOOKUP (Advanced)
Searches both vertical & horizontal

No column number needed

Supports reverse lookup

=XLOOKUP(A2,A2:A10,B2:B10)
XLOOKUP vs INDEX-MATCH
Feature	XLOOKUP	INDEX-MATCH
Ease	Very Easy	Complex
Speed	Fast	Fast
Reverse lookup	Yes	Yes
Recommended	⭐⭐⭐⭐	⭐⭐⭐
Best Choice: XLOOKUP (modern, simpler, powerful)

Sub HighlightStock()
Dim cell As Range
For Each cell In Range("B2:B11")
    If cell.Value < 10 Then
        cell.Interior.Color = vbRed
    ElseIf cell.Value > 50 Then
        cell.Interior.Color = vbGreen
    End If
Next cell
